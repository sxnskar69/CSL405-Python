```python
class conditionalStatements:
    def checkString(self, s: str) -> bool:
        """
        Checks if every 'a' appears before every 'b' in the string.

        Args:
            s: The input string consisting of only 'a' and 'b'.

        Returns:
            True if every 'a' appears before every 'b', False otherwise.
        """

        # Find the index of the first 'b'
        first_b_index = s.find('b')

        # If 'b' is not found, the condition is automatically satisfied
        if first_b_index == -1:
            return True

        # Check if any 'a' appears after the first 'b'
        for i in range(first_b_index, len(s)):
            if s[i] == 'a':
                return False

        return True

obj = conditionalStatements()
print(obj.checkString("aaabbb"))
print(obj.checkString("abab"))
print(obj.checkString("bbb"))
print(obj.checkString("bbbb"))
```
